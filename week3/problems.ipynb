**Constraint Programming**

```python
from ortools.sat.python import cp_model

model = cp_model.CpModel()

x1 = model.NewIntVarFromDomain(cp_model.Domain.FromValues([1, 2]), "x1")
x2 = model.NewIntVarFromDomain(cp_model.Domain.FromValues([1, 2]), "x2")
x3 = model.NewIntVarFromDomain(cp_model.Domain.FromValues([1, 2, 3]), "x3")

model.Add(3 * x1 + x2 + x3 == 10)
model.AddAllDifferent([x1, x2, x3])

solver = cp_model.CpSolver()
status = solver.Solve(model)

if status in (cp_model.OPTIMAL, cp_model.FEASIBLE):
    print("x1 =", solver.Value(x1))
    print("x2 =", solver.Value(x2))
    print("x3 =", solver.Value(x3))
else:
    print("No hay solución")
```

**Latin Square**

```python
n = 4

model = cp_model.CpModel()

grilla = []
for i in range(n):
    fila = []
    for j in range(n):
        fila.append(model.NewIntVar(1, n, "x" + str(i) + str(j)))
    grilla.append(fila)

for i in range(n):
    model.AddAllDifferent(grilla[i])

for j in range(n):
    c = [grilla[i][j] for i in range(n)]
    model.AddAllDifferent(c)

for j in range(n):
    model.Add(grilla[0][j] == j + 1)
for i in range(n):
    model.Add(grilla[i][0] == i + 1)

solver = cp_model.CpSolver()
status = solver.Solve(model)

if status in (cp_model.OPTIMAL, cp_model.FEASIBLE):
    for i in range(n):
        for j in range(n):
            print(solver.Value(grilla[i][j]), end=" ")
        print()
else:
    print("No hay solución")
```

**Sudoku**

```python
n = 9

model = cp_model.CpModel()

grilla = []
for i in range(n):
    fila = []
    for j in range(n):
        fila.append(model.NewIntVar(1, n, "x" + str(i) + str(j)))
    grilla.append(fila)

for i in range(n):
    model.AddAllDifferent(grilla[i])

for j in range(n):
    c = [grilla[i][j] for i in range(n)]
    model.AddAllDifferent(c)

for bloque_fila in range(3):
    for bloque_col in range(3):
        bloque = []
        for i in range(3):
            for j in range(3):
                bloque.append(grilla[bloque_fila * 3 + i][bloque_col * 3 + j])
        model.AddAllDifferent(bloque)

pistas = [
    (1, 9, 9),
    (9, 1, 1),
]

for r, c, v in pistas:
    model.Add(grilla[r - 1][c - 1] == v)

solver = cp_model.CpSolver()
status = solver.Solve(model)

if status in (cp_model.OPTIMAL, cp_model.FEASIBLE):
    for i in range(n):
        for j in range(n):
            print(solver.Value(grilla[i][j]), end=" ")
        print()
else:
    print("No hay solución")
```
